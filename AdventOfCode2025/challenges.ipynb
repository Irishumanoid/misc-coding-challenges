```python
def is_invalid(id: str):
    return len(id) % 2 == 0 and id[:len(id) // 2] == id[len(id) // 2:]

def is_also_invalid(id: str):
    for rep_len in range(1, len(id) // 2 + 1):
        if len(id) % rep_len == 0:
            index = 0
            is_invalid = True
            while index + rep_len < len(id):
                if id[index:index + rep_len] != id[index + rep_len: index + 2 * rep_len]:
                    is_invalid = False
                index += rep_len
            if is_invalid:
                return True
    return False

with open("d2.txt") as f:
    id_ranges = f.read().split(",")


invalid_id_sum = 0
for r in id_ranges:
    seg = r.split("-")
    for i in range(int(seg[0]), int(seg[1]) + 1):
        if is_also_invalid(str(i)):
            invalid_id_sum += i
invalid_id_sum
```

```python
with open("d3.txt") as f:
    banks = f.read().split('\n')

sum = 0
for bank in banks:
    batteries = [int(bank[i]) for i in range(len(bank))]
    first = max(batteries)
    if batteries.index(first) == len(batteries)-1:
        batteries.remove(first)
        second = first
        first = max(batteries)
    else:
        second = max(batteries[batteries.index(first)+1:])
    sum += first*10 + second
sum
```

```python
PAPER = '@'

inc = lambda i, j, data : 1 if data[i][j] == PAPER else 0

def is_accessible(i, j, data):
    sum = 0
    if i-1 >= 0:
        sum += inc(i-1, j, data)
        if j-1 >= 0: sum += inc(i-1, j-1, data)
        if j+1 < len(data[0]): sum += inc(i-1, j+1, data)
    if i+1 < len(data):
        sum += inc(i+1, j, data)
        if j-1 >= 0: sum += inc(i+1, j-1, data)
        if j+1 < len(data[0]): sum += inc(i+1, j+1, data)
    if j-1 >= 0: sum += inc(i,j-1, data)
    if j+1 < len(data[0]): sum += inc(i,j+1, data)
    return sum < 4

with open('d4.txt') as f:
    data = f.read().split('\n')

num = 0
arr_data = []
for i in range(len(data)):
    arr_data.append([])
    for j in range(len(data[i])):
        arr_data[i].append(data[i][j])
# while loop to iterate until no more changes for entire iteration
while True:
    changed = False
    for i in range(len(arr_data)):
        for j in range(len(arr_data[0])):
            if arr_data[i][j] == PAPER and is_accessible(i, j, arr_data):
                changed = True
                num += 1
                arr_data[i][j] = '.'
    if not changed:
        break
num
```

```python
with open("d5.txt") as f:
    data = f.read().split('\n')
    ranges = []
    ids = []
    split = False
    for line in data:
        if line == "":
            split = True
        else:
            if not split:
                ranges.append(line)
            else:
                ids.append(int(line))

sorted_ranges = []
for spread in ranges:
    r = spread.split("-")
    sorted_ranges.append((int(r[0]), int(r[1])))
sorted_ranges.sort(key=lambda r : r[0])

dist_ranges = []
i = 0
start = sorted_ranges[0][0]
end = sorted_ranges[0][1]
while i < len(sorted_ranges) - 1:
    if sorted_ranges[i+1][0] - 1 <= sorted_ranges[i][1]:
        end = max(sorted_ranges[i][1], sorted_ranges[i+1][1])
    else:
        dist_ranges.append((start, end))
        start = sorted_ranges[i+1][0]
        end = sorted_ranges[i+1][1]
    i += 1
dist_ranges.append((start, end))

count = 0
for id in ids:
    fresh = False
    for r in dist_ranges:
        if id >= r[0] and id <= r[1]:
            fresh = True
            break
    if fresh:
        count += 1

num_valid = 0
for r in dist_ranges:
    num_valid += r[1] - r[0] + 1
num_valid
```

```python
import re
import numpy as np

ADD = '+'
MULT = '*'

with open("test.txt") as f:
    data = f.read().split('\n')
vals = []
for i in range(len(data) - 1):
    vals.append([int(v) for v in re.split("\\s+", data[i])])
ops = re.split("\\s+", data[-1])

def num_digits(n):
    count = 1
    while n // 10 > 0:
        count += 1
        n = n // 10
    return count

def get_sum(input):
    nums = input.copy()
    m = max([num_digits(n) for n in nums])
    sum = 0
    for i in range(m):
        cur = 0
        for j in range(len(nums)):
            if m - i <= num_digits(nums[j]):
                cur = cur*10 + (nums[j] % 10)
                nums[j] = nums[j] // 10
        sum += cur
    return sum

def get_prod(input):
    nums = input.copy()
    m = max([num_digits(n) for n in nums])
    prod = 1
    for _ in range(m):
        cur = 0
        for j in range(len(nums)):
            if nums[j] != 0:
                cur = cur*10 + (nums[j] % 10)
                nums[j] = nums[j] // 10
        prod *= cur
    return prod

first_sum = 0
read_sum = 0
for i in range(len(vals[0])):
    v = [val[i] for val in vals]
    if ops[i] == ADD:
        first_sum += np.sum(v)
        read_sum += get_sum(v)
    elif ops[i] == MULT:
        first_sum += np.prod(v)
        read_sum += get_prod(v)
read_sum
```
