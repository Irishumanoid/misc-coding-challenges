```python
# Problem 14 - Longest Collatz Sequence
def collatz(n):
    if n == 1:
        return 0
    elif n % 2 == 0:
        return 1 + collatz(n // 2)
    else:
        return 2 + collatz((3 * n + 1) // 2)

max([(n, collatz(n)) for n in range(500_000, 1_000_000)], key=lambda t : t[1])
```

```python
# Problem 3 - Largest Prime Factor
import numpy as np

def is_prime(num):
    if num == 1: 
        return False
    for i in range(2, int(np.sqrt(num)) + 1):
        if num % i == 0:
            return False
    return True

n = 600851475143
largest_prime = -1
for i in range(2, int(np.sqrt(n)) + 1):
    if is_prime(i) and n % i == 0:
        largest_prime = i
largest_prime
```

```python
# Problem 4 - Largest Palindrome Product
def is_palindrome(num):
    str_num = str(num)
    i = 0
    while i < int(len(str_num) / 2):
        if str_num[i] != str_num[len(str_num) - i - 1]:
            return False
        i += 1
    return True

max = 999
largest = 0
for i in range(max - 99):
    for j in range(i, max + 1):
        cur = (999 - i) * (999 - j)
        if is_palindrome(cur) and cur > largest:
            largest = cur
largest
```

```python
# Problem 9 - Special Pythagorean Triplet

# pythagorean triplet s.t. a+b+c=1000
# let c = 1000 to start so a = sqrt(c^2 - b^2) and iterate through values of b starting from 1 until get 3 integral a and b
sum = 1000
found = False

for a in range(1, sum):
    for b in range(1, a+1):
        c = sum - a - b
        if a ** 2 + b ** 2 == c ** 2:
            print("({}, {}, {}) with product {}".format(a, b, c, a*b*c))
            found = True
            break
    if found:
        break
```

```python
# Problem 10 - Summation of Primes (with Sieve of Eratosthenes)
sum = 0
limit = 2_000_000
cross_lim = int(np.sqrt(limit))
sieve = [False] * (limit - 1)
for i in range(len(sieve)):
    if not sieve[i]:
        p = i + 2
        if p > cross_lim:
            break
        mul = 2
        while p * mul <= limit:
            sieve[p * mul - 2] = True
            mul += 1

for i in range(len(sieve)):
    if not sieve[i]:
        sum += i + 2
sum
```

```python
# Problem 11 - Largest Product in a Grid
import numpy as np
with open("p11.txt") as f:
    lines = f.read().split('\n')
    num_grid = np.zeros((len(lines), len(lines[0].split(" "))))
    for i in range(len(lines)):
        num_grid[i, :] = [int(n) for n in lines[i].split(" ")]

max_prod = 0
r = len(num_grid)
c = len(num_grid[0])
num_terms = 4

def reassign(prod, max_prod):
    return prod if prod > max_prod else max_prod

for i in range(r):
    for j in range(c):
        if j <= c - num_terms:
            max_prod = reassign(np.prod([num_grid[i, j + k] for k in range(num_terms)]), max_prod)
            if i <= r - num_terms:
                max_prod = reassign(np.prod([num_grid[i + k, j + k] for k in range(num_terms)]), max_prod)
        if i <= r - num_terms:
            max_prod = reassign(np.prod([num_grid[i + k, j] for k in range(num_terms)]), max_prod)
            if j >= num_terms - 1:
                max_prod = reassign(np.prod([num_grid[i + k, j - k] for k in range(num_terms)]), max_prod)

print(int(max_prod))
```

```python
# Problem 12 - Highly Divisible Triangular Number

def get_num_divisors(num):
    n = num * (num + 1) / 2
    prime_counts = []
    prod = 1
    for i in range(2, int(n)+1):
        if is_prime(i):
            n_factors = 0
            cur = n
            while cur % i == 0:
                n_factors += 1
                cur /= i
            if n_factors > 0:
                prod *= i ** n_factors
                #print("n={}, p={}, num factors of p={}".format(n, i, n_factors))
            prime_counts.append(n_factors)
            if prod == n:
                break
    num = 1
    for p in prime_counts:
        num *= (p + 1)
    return num


thresh = 500
num = 2
while get_num_divisors(num) < 500:
    num += 1
num
```

```python
# Problem 13 - Large Sum
import numpy as np

# sum the cur last digits and // 10 and add to one place value to the left until reach 0th loc
with open("p13.txt") as f:
    num_strs = f.read().split("\n")

digits = len(num_strs[0])
pv_sums = np.zeros(digits)
pv = 0
extra = 0
while pv < digits:
    s = np.sum([int(n[digits - pv - 1]) for n in num_strs]) + extra
    if pv == digits - 1:
        pv_sums[pv] = s
    else:
        pv_sums[pv] = s % 10
        extra = s // 10
    pv += 1

digits = str(int(pv_sums[-1]))
count = len(digits)
for i in range(10 - count):
    digits += str(int(pv_sums[-i-2]))
int(digits)
```

```python
# Problem 31 - Coin Sums 
# 1p, 2p, 5p, 10p, 20p, 50p, £1 (100p), and £2 (200p). - how many ways to sum to 200p? -- recursive backtracking
options = [1, 2, 5, 10, 20, 50, 100, 200]

def get_num_ways(sum):
    return get_num_ways_rec(sum, 0, [], 0)

def get_num_ways_rec(sum, num_ways, choices: list, cur_min_index):
    if sum == 0:
        num_ways += 1
    elif sum > 0:
        for i in range(cur_min_index, len(options)):
            choices.append(options[i])
            num_ways = get_num_ways_rec(sum - options[i], num_ways, choices, max(cur_min_index, i)) 
            choices.remove(options[i])
    return num_ways

print("num ways: {}".format(get_num_ways(200)))
```

```python
# Problem 28 - Number Spiral Diagonals
import numpy as np

dirs = [(0, 1), (1, 0), (0, -1), (-1, 0)]
count = 0
prev_len = 0
count_len = 1
d = 0

s = 1001
spiral = np.zeros((s, s))
center = (len(spiral) - 1) // 2
cur_x = center
cur_y = center

for i in range(1, len(spiral) * len(spiral[0]) + 1):
    spiral[cur_x, cur_y] = i
    dir = dirs[d % 4]
    cur_x += dir[0]
    cur_y += dir[1]

    count += 1
    if count == count_len:
        d += 1
        if count_len == prev_len:
            count_len += 1
        else:
            prev_len = count_len
        count = 0
sum = 0
print(spiral)
for i in range(len(spiral)):
    sum += spiral[i, i]
    sum += spiral[i, len(spiral[i]) - i - 1]
sum -= spiral[center][center]
print(int(sum))
```

```python
# Problem 43 - Sub-string Divisibility
def is_valid(s):
    return (int(s[1:4]) % 2  == 0 and
            int(s[2:5]) % 3  == 0 and
            int(s[3:6]) % 5  == 0 and
            int(s[4:7]) % 7  == 0 and
            int(s[5:8]) % 11 == 0 and
            int(s[6:9]) % 13 == 0 and
            int(s[7:10]) % 17 == 0)

def get_valid_permutation_sum(num_str: str, cur_str: str, sum):
    if len(num_str) == 0:
        sum += int(cur_str) if is_valid(cur_str) else 0
    for i in range(len(num_str)):
        next_cur = cur_str + num_str[i]
        next_num_str = num_str[:i] + num_str[i+1:]
        sum = get_valid_permutation_sum(next_num_str, next_cur, sum)
    return sum

print(get_valid_permutation_sum("0123456789", "", 0))
```

```python
# Problem 24 - Lexicographic Permutations

factorial = lambda n : 1 if n == 0 or n == 1 else n * factorial(n - 1)

count = 1_000_000
p_str = ""
nums = list(range(10))

count -= 1
while len(nums) > 0:
    factorial_rem = factorial(len(nums) - 1)
    digit_index = count // factorial_rem
    p_str += str(nums[digit_index])
    nums.pop(digit_index)
    count %= factorial_rem
print(p_str)
```

```python
# Problem 18 - Maximum Path Sum I
with open("p18.txt") as f:
    str_rows = f.read().split("\n")
    rows = []
    for r in str_rows:
        rows.append([int(n) for n in r.split(" ")])

max_sum = 0
for i in range(2 ** (len(rows) - 1)):
    binary = bin(i)[2:]
    binary += "0" * (len(rows) - len(binary) - 1)
    index = 0
    sum = rows[0][0]
    for j in range(1, len(rows)):
        index = index + 1 if int(binary[j - 1]) == 1 else index
        sum += rows[j][index]
    if sum > max_sum:
        max_sum = sum
print("max sum: {}".format(max_sum))
```

```python
# Problem 23 - Non-Abundant Sums
import numpy as np

def num_factors(n, k):
    num = 0
    while n % k == 0:
        num += 1
        n /= k
    return num
# find all nums that are a sum of 2 abundant nums (find these) and subtract from total sum
def is_abundant(n):
    if is_prime(n):
        return False
    p_and_e = [] # primes and exponents - list of tuples
    for i in range(2, n):
        if is_prime(i) and n % i == 0:
            p_and_e.append((i, num_factors(n, i)))
    factor_sum = np.prod([(t[0] ** (t[1] + 1) - 1)/(t[0] - 1) for t in p_and_e]) - n
    return factor_sum > n

all = []
for i in range(12, 28124):
    if is_abundant(i):
        all.append(i)

def is_abundant_sum(num, all):
    for entry in all:
        if entry > num:
            return False
        elif num - entry in all:
            return True
    return False
```

```python
sum = 0
for i in range(1, 28124):
    if not is_abundant_sum(i, all):
        sum += i
sum
```
